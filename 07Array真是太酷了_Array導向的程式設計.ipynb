## 2. Dot Product — AI 最愛的運算

> 向量內積（dot product）是 Google 搜尋、推薦系統、ChatGPT 注意力機制的核心！

數學定義：

$$\mathbf{u} \cdot \mathbf{v} = u_1 v_1 + u_2 v_2 + u_3 v_3$$

幾何意義：**兩個向量有多「像」**

```python
u = np.array([1, 2, 3])
v = np.array([1, -1, 0])
```

```python
# 手算
np.sum(u * v)
```

```python
# 這麼重要的動作，numpy 當然有方便的符號！
u @ v
```

### 矩陣

一個 $2 \times 3$ 的矩陣 $A$，一列一列打進去：

$$A=\begin{bmatrix}1 & 2 & 3 \\ 3 & 2 & 1 \end{bmatrix}$$

```python
A = np.array([[1, 2, 3],
              [3, 2, 1]])
```

```python
A[0, 1]  # 比 list 的 A[0][1] 更像數學
```

$$B = \begin{bmatrix}1 & 2 \\ 2 & 1 \\ 1 & 1 \end{bmatrix}$$

```python
B = np.array([[1, 2],
              [2, 1],
              [1, 1]])

print('A.shape:', A.shape)
print('B.shape:', B.shape)
```

$$C = AB$$

每個位置是「前面出列」乘「後面出行」的內積。

```python
C = A @ B
print('C.shape:', C.shape)
C
```

### 矩陣最重要的事！

向量乘矩陣 = **列向量的線性組合**

這是 AI 的核心運算模式：把輸入看成係數，矩陣看成「基底」，輸出是它們的組合。

```python
x = np.array([1, 2, -1])
x @ B  # 1*[1,2] + 2*[2,1] + (-1)*[1,1]
```

## 3. 推薦系統 — 用矩陣做出來！

這不是玩具，Netflix、Spotify、YouTube 都用類似的概念！

想像每個商品可以用一個向量描述它的「特徵」：

| 商品 | 特徵1 | 特徵2 | 特徵3 |
|------|-------|-------|-------|
| A    | 4     | 7     | 2     |
| B    | 1     | 2     | 3     |
| C    | 2     | 9     | 7     |
| D    | 5     | 2     | 0     |
| E    | 1     | 7     | 4     |

```python
# P 矩陣：每一列是一個商品的「特徵向量」
P = np.array([[4, 7, 2],
              [1, 2, 3],
              [2, 9, 7],
              [5, 2, 0],
              [1, 7, 4]])

print('P.shape:', P.shape)  # 5 個商品，每個有 3 個特徵
```

```python
# 使用者 u 的偏好向量（對 3 個特徵的喜好程度）
u = np.array([4, 8, 7])
```

用 dot product 計算這位使用者對每個商品的「相關分數」：

分數越高 = 這位使用者越可能喜歡這個商品！

```python
scores = u @ P.T
print('各商品的推薦分數:', scores)
print('最推薦的商品編號:', np.argmax(scores))
```

```python
# 換個使用者試試看
u2 = np.array([3, 0, 4])
scores2 = u2 @ P.T
print('各商品的推薦分數:', scores2)
print('最推薦的商品編號:', np.argmax(scores2))
```

### 🤯 這就是 AI 推薦系統的雛形！

真實的 Netflix/Spotify 做的事情也差不多：
1. 把每部電影/每首歌用向量表示（稱為 **embedding**）
2. 把每個使用者也用向量表示
3. 用 dot product 算分數
4. 推薦分數最高的內容

ChatGPT 的「注意力機制」也是一堆 dot product！

## 4. Tensor — AI 數據的基本單位

在深度學習裡，每一筆數據都是一個 **tensor**。

```python
a = 87                                          # 純量 (0 階 tensor)
b = np.array([4, 8, 7])                          # 向量 (1 階 tensor)
c = np.array([[9, 4], [8, 7]])                   # 矩陣 (2 階 tensor)
d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])  # 3D (3 階 tensor)

print('b.shape:', b.shape)  # 1D
print('c.shape:', c.shape)  # 2D
print('d.shape:', d.shape)  # 3D
```

### 日常 tensor 的例子：

| Tensor 階數 | 例子 |
|------------|------|
| 0 階（純量）| 一個數字：溫度 37.2°C |
| 1 階（向量）| 一筆特徵：[身高, 體重, 年齡] |
| 2 階（矩陣）| 一張灰階圖片（像素值） |
| 3 階 | 一張彩色圖片（RGB 三層） |
| 4 階 | 一批彩色圖片（batch × H × W × C） |

## 5. Softmax — AI 的「機率化」魔法

AI 做分類的最後一步，常常是把一堆分數「變成機率」。

比如分辨貓/狗/鳥，模型給出分數 [1.9, 1.1, -3]，
Softmax 把它們轉成加起來等於 1 的機率！

$$\text{softmax}(x_i) = \dfrac{e^{x_i}}{\sum_j e^{x_j}}$$

```python
def softmax87(x):
    """單一向量版 softmax"""
    e_x = np.exp(x)
    return e_x / e_x.sum()

x = np.array([1.9, 1.1, 0.2])
probs = softmax87(x)

print('機率:', probs)
print('加起來:', probs.sum())  # 一定是 1！
```

```python
# 進階版：同時處理多筆數據
def softmax(x):
    """可同時處理多組向量的 softmax"""
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=-1, keepdims=True)

# 兩批分類分數一次搞定
Y = np.array([[1.9, 1.1, 0.2],
              [2.3, 1.7, 3.0]])
softmax(Y)
```

### 視覺化 Softmax

原始分數 vs. 機率分佈

```python
labels = ['貓', '狗', '鳥']
raw_scores = np.array([1.9, 1.1, -3.0])
probs = softmax87(raw_scores)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].bar(labels, raw_scores, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
axes[0].set_title('原始分數')

axes[1].bar(labels, probs, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
axes[1].set_title('Softmax 機率')
axes[1].set_ylim(0, 1)

plt.suptitle('Softmax 的效果', fontsize=14)
plt.tight_layout()
plt.show()
```

## 6. Axis — 運算的方向

`axis=0` 是垂直方向（跨列），`axis=1` 是水平方向（跨欄）。

這在深度學習的 batch 計算裡非常重要！

```python
A = np.array([[1, 2],
              [3, 4]])

print('全部加總:', A.sum())
print('axis=0（每欄加總）:', A.sum(axis=0))
print('axis=1（每列加總）:', A.sum(axis=1))
```

想像成一張成績表：

|       | 數學 | 英文 |
|-------|------|------|
| 同學A | 1    | 2    |
| 同學B | 3    | 4    |

- `axis=0` → 算每科的平均（垂直）
- `axis=1` → 算每人的總分（水平）

```python
print('每科平均:', A.mean(axis=0))
print('每人總分:', A.sum(axis=1))
```

**挑戰 2：溫度 Softmax**

AI 生成文字時有個「溫度」(temperature) 參數，控制輸出的多樣性：

$$\text{softmax}(x / T)$$

- $T$ 接近 0 → 只選最高分（保守）
- $T$ 很大 → 機率接近均勻（有創意）

試著調整 T，觀察機率分佈如何變化！

```python
from ipywidgets import interact

scores = np.array([3.0, 2.0, 1.0, 0.5])  # 4 個選項的分數
labels_ch = ['A', 'B', 'C', 'D']

def plot_softmax_temp(T=1.0):
    probs = softmax87(scores / T)
    plt.bar(labels_ch, probs, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
    plt.ylim(0, 1)
    plt.title(f'Temperature = {T:.1f}')
    plt.ylabel('機率')
    plt.show()

interact(plot_softmax_temp, T=(0.1, 5.0, 0.1));
```
