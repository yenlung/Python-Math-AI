## 2. 試做 softmax 函數

softmax 函數在 AI 世界是很常用的小技巧函數。簡單說是我們給了要分類的, 比方說三個類別, 給了 3 個分數。我們希望三個數字加起來變成 1, 這樣可以視為一個機率分佈。

```python
a = 1.9
b = 1.1
c = -3
```

把 3 個數字都變成正數, 但保持大小關係。

```python
a = np.exp(a)
b = np.exp(b)
c = np.exp(c)
```

轉化加起來等於 1。

```python
s = a + b + c

a = a/s
b = b/s
c = c/s
```

```python
print(a)
print(b)
print(c)
```

```python
def softmax(a, b, c):
    a = np.exp(a)
    b = np.exp(b)
    c = np.exp(c)
    s = a + b + c
    a = a/s
    b = b/s
    c = c/s
    return a, b, c
```

```python
a, b, c = softmax(1.9, 1.1, -3)
```

```python
print(a)
print(b)
print(c)
```

```python
a+b+c
```

## 4. 互動畫圖

```python
def plot_func(a):
    x = np.linspace(-5,5,100)
    y = a*x**2
    plt.plot(x,y)
    plt.ylim(0,50)
    plt.show()

interact(plot_func, a=(0,5));
```

## 5. 函式 × 藝術 × 互動

```python
def art(k):
    t = np.linspace(0,2*np.pi,1000)
    r = np.sin(k*t)
    x = r*np.cos(t)
    y = r*np.sin(t)
    plt.plot(x,y)
    plt.axis('equal')
    plt.show()

interact(art, k=(1,10));
```

## 9. 用等高線畫隱函數

## 10. 超橢圓

$$ |\dfrac{x}{a}|^n + |\dfrac{y}{b}|^n - 1 = 0$$

```python
n=4

x = y = np.linspace(-1.5, 1.5, 200)
X, Y = np.meshgrid(x, y)
Z = np.abs(X)**n + np.abs(Y)**n - 1
plt.axis('equal')
plt.axis('off')
plt.contour(X, Y, Z, 0);
```

## 11. 超橢圓參數式

$$\begin{cases} x = |\cos(\theta)|^{2/n} \times \mathrm{sgn}(\cos(\theta))\\
y = |\sin(\theta)|^{2/n} \times \mathrm{sgn}(\sin(\theta)) \end{cases}$$

```python
π = np.pi
```

```python
θ = np.linspace(0, 2*π, 200)
```

```python
n = 3

x = np.abs(np.cos(θ))**(2/n) * np.sign(np.cos(θ))
y = np.abs(np.sin(θ))**(2/n) * np.sign(np.sin(θ))

plt.axis('equal')
plt.axis('off')
plt.fill(x, y, c='#F86503');
```

```python
def superellipse(n=2):
    π = np.pi
    θ = np.linspace(0, 2*π, 200)
    x = np.abs(np.cos(θ))**(2/n) * np.sign(np.cos(θ))
    y = np.abs(np.sin(θ))**(2/n) * np.sign(np.sin(θ))

    plt.axis('equal')
    plt.axis('off')
    plt.fill(x, y, c='#F86503');

    plt.show()
```

```python
superellipse(3.2)
```

```python
interact(superellipse, n=(2, 20, 0.2));
```
