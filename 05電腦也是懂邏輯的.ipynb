## 3. `while` 迴圈和可以一直聊下去的拍拍機器人

`while True` 是要讓程式一直執行的小技巧。

```python
character = "(*´∀`)~♥"

while True:
    prompt = input("> ")
    reply = "拍拍。"
    print(character + ": " + reply)
    print()
```

#### 看到 Bye 就再見的拍拍機器人

但是要注意大家用法會不一樣, 比如說 "Bye～～", 比如說 "Bye Bye", 或者 "Bye Bye, 我去上課了" 等等...

```python
"Bye" == "Bye～～"
```

可以很像數學的表示, 看某個字串是不是在一個句子中出現。

```python
"Bye" in "Bye～～"
```

愛聊多久聊多久, 直到你說 "Bye" 的拍拍機器人。

```python
character = "(*´∀`)~♥"
egg = ["拍拍。", "秀秀。", "沒關係。"]
flag = True

while flag:
    prompt = input("> ")

    if "Bye" in prompt:
        reply = "好哦, 下次再聊!"
        flag = False
    else:
        reply = np.random.choice(egg, 1)[0]

    print(character + ": " + reply)
    print()
```

## 4. 簡單有趣的應用: 猜數字遊戲

```python
ans = np.random.randint(1, 101)
guess = 9487 # 設一個完全不可能的答案

while guess != ans:
    guess = int(input("請輸入你要猜的數字: "))
    if guess > ans:
        print("太大了!")
    elif guess < ans:
        print("太小了!")
    else:
        print("太神了!")
    print()
```

#### 🎬 表情動畫（邏輯 + 亂數）

```python
from time import sleep
from IPython.display import clear_output
```

```python
for n in range(50):
    clear_output(wait=True)
    x = np.random.rand()

    if x < 0.3:
        print(" "*n + "😴")
    elif x < 0.6:
        print(" "*n + "🙂")
    else:
        print(" "*n + "🤩")

    sleep(0.5)
```

#### 👩‍💻 挑戰: 用不同的方法說再見

其實我們不只用 "Bye", 有更多說再見的方式。

####

```python
bye = ["走了", "再見", "Bye"]
```

```python
bye = ["走了", "再見", "Bye"]
flag = True

prompt = input("> ")

for word in bye:
  if word in prompt:
    flag = False

if flag:
    print("拍拍")
else:
    print("拜拜, 下次再聊")
```

Gemini 教我們好高級的方式。

```python
prompt = "我要先走了，Bye!"
bye = ["走了", "再見", "Bye"]

if any(word in prompt for word in bye):
    print("偵測到離別詞彙")
else:
    print("未偵測到離別詞彙")
```

這是什麼意思呢?

```python
[word in prompt for word in bye]
```
