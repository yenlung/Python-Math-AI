```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. 畫圖的兩大指令

在 Python 的畫圖中，最重要的就是學習 `plt.plot`，·和 `plt.scatter`。

```python
x = [0, 1, 2, 3]
y = [2, -4, 1, 7]
```

之前我們用過，`plt.plot` 就是描點，再把點連起來。

```python
plt.plot(x, y)
```

`plt.scatter` 是散佈圖，也就是只畫出點，不連起來。這在數據圖形呈現上常用。

```python
plt.scatter(x, y)
```

## 2. 參數式其實沒有那麼可怕

畫圖的本質是：算出很多點 (x, y)，然後連起來。

參數式只是改成：

$$\begin{cases} x = f(t) \\ y = g(t)\end{cases}$$

然後讓 $t$ 從某個時間開始，到另一個時間結束。

## 3. 畫一個圓（但 Python 不一定覺得它是圓 🤨）

極座標 $(r, \theta)$ 換成直角座標系統 $(x, y)$ 的方法:

$$\begin{cases} x = r\cos(\theta) \\ y = r\sin(\theta)\end{cases}$$

現在我們試著畫出 $r=3$ 的圓。

```python
t = np.linspace(0, 2*np.pi, 500)
```

```python
r = 3
```

```python
x = r * np.cos(t)
y = r * np.sin(t)
```

```python
plt.plot(x,y);
```

為什麼看起來不像圓？

因為 `matplotlib` 有自己的審美觀。

我們要告訴它：x, y 比例要一樣。

```python
# ax = plt.gca()
# ax.set_aspect('equal')
plt.axis('equal')
plt.plot(x, y)
```

## 4. 補充說明：`matplotlib` 的結構

`matplotlib` 的架構其實是：

	Figure
	 └── Axes (可以很多個)
	        └── plot

* `figure` = 整張圖

* `axes` = 子圖（座標系統）

## 5. 我們可以定義 $\pi$ 真的是 $\pi$!

```python
π = np.pi
```

```python
π
```

```python
θ = np.linspace(0, 2*π, 200)
```

## 6. 笛卡兒的情書 💘

傳說第十三封信只有一行公式：

$$r = 1 - \sin(θ)$$

這到底畫出什麼？

```python
r = 1 - np.sin(θ)
```

```python
x = r * np.cos(θ)
y = r * np.sin(θ)
```

```python
plt.axis('equal')
plt.plot(x, y)
```

❤️ 這就是心形。

## 8. 有意識的創作 🎨

試試看把 $\sin(\theta)$ 改成 $\sin(5\theta)$。

你能預測圖形嗎？

```python
θ = np.linspace(0, 2*π, 1000)
```

```python
r = 1 + 0.3 * np.sin(5*θ)
```

```python
x = r * np.cos(θ)
y = r * np.sin(θ)
```

```python
plt.plot(x, y);
```
