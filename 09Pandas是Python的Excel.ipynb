## 2. 讀入真實數據

從教育部官方網站讀取 112 學年度大專校院學生數統計。

這是貨真價實的公開數據，不是玩具！

```python
url = "https://stats.moe.gov.tw/files/detail/112/112_student.csv"
df = pd.read_csv(url, encoding='utf-8')

print('資料大小:', df.shape)
print('欄位:', list(df.columns))
```

```python
# head() 是 pandas 最常用的指令，看前 5 筆數據
df.head()
```

## 3. 篩選數據：像布林遮罩一樣

先來理解 numpy 的布林篩選，再看 pandas 怎麼做。

```python
egg = np.array([-3, 2, -5, 4, 3, 7, 6, -1])

# 產生布林遮罩
mask = egg > 0
print('布林遮罩:', mask)

# 用遮罩篩選
print('正數:', egg[mask])
```

```python
# pandas 的做法完全一樣！

# 先只留下日間部學士班
df_filtered = df[df['日間∕進修別'] == 'D 日']
df_bachelor = df_filtered[df_filtered['等級別'] == 'B 學士']

print('日間部學士班學校數:', len(df_bachelor))
df_bachelor.head()
```

```python
# 選出我們需要的欄位
selected_columns = ['學校名稱', '總計', '男生計', '女生計',
                    '一年級男', '一年級女', '二年級男', '二年級女',
                    '三年級男', '三年級女', '四年級男', '四年級女',
                    '延修生男', '延修生女']

df = df_bachelor[selected_columns].copy()
df.head()
```

## 4. 計算新欄位與排序

Pandas 可以直接在 DataFrame 上計算新的欄位，就像 Excel 的公式！

```python
# 計算女生佔比
df['女生佔比'] = df['女生計'] / df['總計']
df.head()
```

```python
# 按女生佔比排序（由高到低）
df = df.sort_values(by='女生佔比', ascending=False)

# 重設 index，讓 index 代表排名
df = df.reset_index(drop=True)
df.index = df.index + 1  # 從第 1 名開始

df.head(10)
```

```python
# 政大到底第幾名？
rank = df.index[df['學校名稱'] == '國立政治大學'][0]
ratio = df.loc[rank, '女生佔比']

print(f'國立政治大學女生佔比排名：第 {rank} 名')
print(f'女生佔比：{ratio:.1%}')
```

### 🤔 發現了什麼？

政大並不是第一名！前幾名是哪些學校呢？讓我們畫圖來看看。

```python
df_top20 = df.head(20)

plt.figure(figsize=(12, 6))
sns.barplot(x='學校名稱', y='女生佔比', data=df_top20, palette='RdPu')

plt.title('女生比例最高前 20 名大學（112 學年度日間部學士班）', fontsize=14)
plt.xlabel('學校名稱')
plt.ylabel('女生佔比')
plt.xticks(rotation=45, ha='right')

# 加一條 50% 的參考線
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7, label='50%')
plt.legend()
plt.tight_layout()
plt.show()
```

### 幾所知名大學比較

```python
schools = ['國立政治大學', '國立清華大學', '國立陽明交通大學', '國立中央大學', '國立臺灣大學']
df_compare = df[df['學校名稱'].isin(schools)][['學校名稱', '總計', '女生佔比']]
df_compare
```

```python
plt.figure(figsize=(8, 5))
colors = ['#FF6B6B' if ratio > 0.5 else '#4ECDC4'
          for ratio in df_compare['女生佔比']]

bars = plt.bar(df_compare['學校名稱'], df_compare['女生佔比'], color=colors)
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7)

# 在每個長條上顯示百分比
for bar, ratio in zip(bars, df_compare['女生佔比']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{ratio:.1%}', ha='center', va='bottom', fontsize=10)

plt.title('幾所知名大學女生佔比比較')
plt.ylabel('女生佔比')
plt.xticks(rotation=15)
plt.ylim(0, 0.8)
plt.tight_layout()
plt.show()
```

## 5. Seaborn 圖表展示廳

Seaborn 有很多很厲害的視覺化方式，以下是幾個經典範例。

這些都是用 AI（ChatGPT/Gemini）協作生成的——看完之後，想想你想深入哪一種！

```python
from seaborn import load_dataset

# 載入 iris 鳶尾花資料集（機器學習的 Hello World）
iris = load_dataset('iris')
tips = load_dataset('tips')

print('iris 資料集:', iris.shape)
iris.head()
```

### Pairplot：一次看所有特徵之間的關係

資料科學家最愛用的「一鍵探索」工具！

```python
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle('鳶尾花各特徵兩兩比較', y=1.02)
plt.show()
```

### Heatmap：相關係數矩陣

一眼看出哪些特徵「互相有關」——這在 AI 特徵選取時很重要！

```python
plt.figure(figsize=(6, 5))
corr = iris.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('鳶尾花特徵相關係數矩陣')
plt.tight_layout()
plt.show()
```

### Violin Plot：分佈的形狀

比 boxplot 更漂亮，也更能看出資料的分佈形狀。

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='species', y='petal_length', data=iris, palette='Set2')
plt.title('各品種花瓣長度分佈')
plt.xlabel('品種')
plt.ylabel('花瓣長度（cm）')
plt.show()
```

### Scatter + 分群：小費數據探索

用不同顏色、大小來表現多個維度的資訊。

```python
sns.relplot(data=tips, x='total_bill', y='tip',
            hue='day', size='size', kind='scatter')
plt.suptitle('帳單金額 vs 小費（不同天、不同桌人數）', y=1.02)
plt.show()
```

### Regression Plot：一秒畫出回歸線

帳單金額和小費有關係嗎？男女給的比例一樣嗎？

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex',
           height=5, aspect=1.2)
plt.title('帳單 vs 小費（含回歸線）')
plt.show()
```

### KDE Plot：二維密度圖

資料集中在哪個區域？

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=iris, x='sepal_width', y='sepal_length',
            hue='species', fill=True)
plt.title('花萼寬度 vs 花萼長度（密度圖）')
plt.show()
```

# 🧠 核心觀念回顧

| 操作 | 程式碼 | 說明 |
|------|--------|------|
| 讀 CSV | `pd.read_csv(url)` | 從網路或本地讀數據 |
| 看數據 | `df.head()` | 看前幾筆 |
| 篩選 | `df[df['欄位'] == 值]` | 布林遮罩 |
| 新欄位 | `df['新欄'] = ...` | 像 Excel 公式 |
| 排序 | `df.sort_values(by='欄位')` | 升冪/降冪 |
| 重設索引 | `df.reset_index(drop=True)` | 重新編號 |
